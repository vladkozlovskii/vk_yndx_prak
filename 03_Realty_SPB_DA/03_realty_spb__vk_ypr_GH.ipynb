один потолок высотой в 100 метров (после этого и решил смотреть), кроме того еще целый ряд значений вокург 20 м (в районе 20 м это может быть связано с ошибкой в знаке), но потолки в районе 5 м в каких-то отдельных квартирах могут и быть + очень странно выглядел потолок в 1 м (при первом взгляде на данные), число потолков меньше 2 м тоже стоит посмотреть

### Обработка ceiling_height

```python
# обработка столбца ceiling_height
ceiling_median = data['ceiling_height'].median()
print('медианная высота потолка:', ceiling_median) # медианная выстоа потолка (до обработки)
def ceiling_corr(ceiling):     # функция для корректировки аномальных значений высоты потолка
    if ceiling > 40:           # сверханомлаьные значения сводятся к медиане 
        return ceiling_median 
    if ceiling >= 20:           # значения между 20 и 40 делятся на 10
        return ceiling / 10
    if ceiling >= 6:            # значения между 6 и 20 сводятся к медиане (тоже аномалии, но не в результате умнолжения на 10)
        return ceiling_median
    if ceiling >= 1.8:          # нормальные значения осатются такими же
        return ceiling
    if ceiling < 1.8:           # аномально малые значения сводятся к медиане
        return ceiling_median
    
data['ceiling_height'] = data['ceiling_height'].apply(ceiling_corr) # применение функции
data['ceiling_height'].plot(style='o') # график для проверки
```

```python
# заполнение пропусков
ceiling_median = data['ceiling_height'].median() # пересчет медианного значения
data['ceiling_height'] = data['ceiling_height'].fillna(ceiling_median) # замена пропусков медианным значением
print(data['ceiling_height'].isna().sum()) # проверка пропусков
```

```python
def price_cat(price):     # функция для категоризации цен (для использования в графиках)
    if price > 20000000:           # сверхвысокие значения 
        return 6 
    if price >= 11500000:           # то что от границв верхнего уса до границы сверхвысоких значений
        return 5
    if price >= 6800000:            # то что от Q3 до границы верхнего уса
        return 4
    if price >= 3400000:          # то что в границиах IQR
        return 3
    if price >= 1000000:           # от миллиона до Q1
        return 2
    else:                         # меньше миллиона
        return 1
    
data['price_cat'] = data['last_price'].apply(price_cat) # применение функции
data.plot(                                              # гистоограмма для проверки
    kind='hist',
    y='price_cat',
    bins=11,
    grid=True,
    legend=True
) 
```

- от 11 500 000 до 20 000 000:            то что от границв верхнего уса до границы сверхвысоких значений - группа 5
- от 6 800 000 до 11 500 000:            то что от Q3 до границы верхнего уса  - группа 4   
- от 3 400 000 до 6 800 000:             то что в границиах IQR  - группа 3

- < 1 000 000:                            - группа 1

**зависимость цены от этажа**

```python
def floors_categorized_d(floor):  # функция для перевода текстовых категорий этажей в числовые
    if floor == 'одноэтажный':
        return 0
    if floor == 'первый этаж':
        return 1    
    if floor == 'другой':
        return 2
    if floor == 'последний этаж':
        return 3    
data['floors_categorized_d'] = data['floors_categorized'].apply(floors_categorized_d) # применение функции
```

```python
data.plot(                    # визуализация зависимости цены и категорий этажа
    kind='scatter',
    y='last_price',
    x='floors_categorized_d',
    grid=True,
    legend=True
)
```

```python
sns.catplot(x = 'floors_categorized_d', y = 'last_price', data = data) # то же через catplot в seaborn - чуть лучше видно, но не смог сетку добавить
plt.show()
# барплот показывающий изменение цены для разных категорий этажей
sns.barplot(x = 'floors_categorized_d', y = 'last_price', ci = None, data = data)

# ящики с усами для той же цели
sns.catplot(x = 'floors_categorized_d', y = 'last_price', kind='box', data = data)
```

Видно, что одноэтажные дома - самые дешевые.
