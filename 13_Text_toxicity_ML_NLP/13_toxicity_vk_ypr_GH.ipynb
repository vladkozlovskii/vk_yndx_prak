ниже идет функция для обработки текста она в основном сходна с процессом в тренажере (добавил еще перевод в нижний регистр, лемматизация идет через WorldLemmatize - т.к. текст английский, далее остаются только буквы и апостроф, далее выдается слитный текст

```python
lemmatizer = WordNetLemmatizer()
nltk.download('omw-1.4')
```

```python
def filtration(text):
  text = text.lower()
  text = ''.join(lemmatizer.lemmatize(text))
  text = re.sub(r'[^a-z\']', ' ', text)
  text = text.split()
  
  return ' '.join(text)
```

```python
%%time
data['text_proc'] = data['text'].apply(filtration)
```

```python
%%time
def f1_score_(target_test, predictions):
    recall = recall_score(target_test, predictions)
    precision = precision_score(target_test, predictions)
    return 2 * (precision * recall) / (precision + recall)
```

```python
f1_scorer = make_scorer(f1_score_, greater_is_better=True)
```

```python
# сделаю списки, чтобы заносить время выполнения и f1 для того чтобы вывести в таблице
score_list = []
time_list = []
```

```python
print('F1_score=', f1_score_(target_test, prediction_logit))
```

##### DecisionTree
Дальше хочу посмотреть, что получится при применении Дерева решений. Для проверки гиперпараметра max_depth хочу использовать вариант функции, которую делал несколько проектов назад - в принципе я понимаю, что более оптимальным может быть использование GridSearchCV - но пока так и не понял как там вывести график

```python
# функция будет принимать признаки и целевой признак, диапазон деревьев и число оценок в кросс-валидации
def tree_params(features, target, range_depth, number_cv):

    best_f1 = 0        # переменная для сохранения лучшего значения f1 
    best_depth_f1 = 0  # переменная для сохранения лучшей глубины дерева при этом

    depth_list = []    # список для записи глубины деревьев
    f1_list = []       # список для записи rmse соответствующего глубине

       
    for depth in range_depth:
        print('*') # чтобы убедиться, что компьютер не завис
        model = DecisionTreeClassifier(random_state=12345, max_depth=depth, class_weight='balanced')
        f1_scores = cross_val_score(model, features, target, scoring=f1_scorer, cv=number_cv, verbose=2)   # кросс-валидация
        final_f1_score = np.abs(f1_scores.mean())                              # cредняя оценка качества модели
        print(f1_scores)
     
        depth_list.append(depth)            # добаление значения глубины дерева в список для графика
        f1_list.append(final_f1_score)          # добаление значения f1 в список для графика
         
        if final_f1_score > best_f1:   # выцепление лучшего значения f1 и соответствующей глубины дерева и числа деревьев
            best_f1 = final_f1_score
            best_depth_f1 = depth
            
 
 
    print('Для модели лучшее значение f1 при кросс-валидации:', best_f1, 'при этом глубина дерева:', best_depth_f1)

    plt.figure(figsize=(15, 6))
    sns.set(context = 'notebook', style = 'whitegrid', font_scale=1)
                        

    sns.scatterplot(x=depth_list, y=f1_list)  # диаграмма рассеяния f1 в зависимости от глубины
    plt.xlabel('Число деревьев', fontsize=12)
    plt.ylabel('f1 score', fontsize=12)
    plt.title('f1 в зависимости от числа деревьев')
```

```python
model_class, tokenizer_class, pretrained_weights = (transformers.DistilBertModel, transformers.DistilBertTokenizer, 'distilbert-base-uncased')
```

```python
tokenizer = tokenizer_class.from_pretrained(pretrained_weights)
model = model_class.from_pretrained(pretrained_weights)
```

```python
%%time
tokenized = sample['text_proc'].apply(lambda x: tokenizer.encode(x,truncation=True , max_length = 100, add_special_tokens=True)) 
```

```python
%%time
padded = np.array([i + [0]*(100 - len(i)) for i in tokenized.values])
```

```python
%%time
attention_mask = np.where(padded != 0, 1, 0)
```

```python
attention_mask.shape
```

```python
%%time
batch_size = 100
embeddings = []
for i in notebook.tqdm(range(padded.shape[0] // batch_size)):
        batch = torch.LongTensor(padded[batch_size*i:batch_size*(i+1)]) 
        attention_mask_batch = torch.LongTensor(attention_mask[batch_size*i:batch_size*(i+1)])
        
        with torch.no_grad():
            batch_embeddings = model(batch, attention_mask=attention_mask_batch)
        
        embeddings.append(batch_embeddings[0][:,0,:].numpy())

        del batch
        del attention_mask_batch
        del batch_embeddings
```

```python
features = np.concatenate(embeddings)
print(features.shape)
```

```python
# разделение на выборки после подготовки признаков с BERT
features_train,  features_test, target_train, target_test = train_test_split(        
        features, target, test_size=0.25, random_state=12345)
print(features_train.shape)
print(target_train.shape)
print(features_test.shape)
print(target_test.shape)
```
